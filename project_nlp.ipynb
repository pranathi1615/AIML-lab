```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, LSTM, Embedding, Dense
from tensorflow.keras.callbacks import EarlyStopping
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction
import warnings
import os
warnings.filterwarnings('ignore')

# Install required libraries
print("Installing required libraries...")
os.system('pip install openpyxl xlrd -q')

# ROUGE calculation functions
def calculate_rouge_n(reference, candidate, n=1):
    """Calculate ROUGE-N score"""
    ref_tokens = reference.lower().split()
    cand_tokens = candidate.lower().split()

    if len(ref_tokens) < n or len(cand_tokens) < n:
        return 0.0

    # Create n-grams
    ref_ngrams = [tuple(ref_tokens[i:i+n]) for i in range(len(ref_tokens)-n+1)]
    cand_ngrams = [tuple(cand_tokens[i:i+n]) for i in range(len(cand_tokens)-n+1)]

    if len(ref_ngrams) == 0:
        return 0.0

    # Count matches
    matches = sum(1 for ngram in cand_ngrams if ngram in ref_ngrams)

    # Calculate recall-based F1
    recall = matches / len(ref_ngrams) if len(ref_ngrams) > 0 else 0
    precision = matches / len(cand_ngrams) if len(cand_ngrams) > 0 else 0

    if recall + precision == 0:
        return 0.0

    f1 = 2 * (precision * recall) / (precision + recall)
    return f1

def calculate_rouge_l(reference, candidate):
    """Calculate ROUGE-L score using LCS"""
    ref_tokens = reference.lower().split()
    cand_tokens = candidate.lower().split()

    if len(ref_tokens) == 0 or len(cand_tokens) == 0:
        return 0.0

    # LCS calculation
    m, n = len(ref_tokens), len(cand_tokens)
    dp = [[0] * (n + 1) for _ in range(m + 1)]

    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if ref_tokens[i-1] == cand_tokens[j-1]:
                dp[i][j] = dp[i-1][j-1] + 1
            else:
                dp[i][j] = max(dp[i-1][j], dp[i][j-1])

    lcs_length = dp[m][n]

    # Calculate F1
    recall = lcs_length / len(ref_tokens) if len(ref_tokens) > 0 else 0
    precision = lcs_length / len(cand_tokens) if len(cand_tokens) > 0 else 0

    if recall + precision == 0:
        return 0.0

    f1 = 2 * (precision * recall) / (precision + recall)
    return f1

# Load dataset with multiple fallback methods
print("="*80)
print("LOADING DATASET")
print("="*80)

file_path = '/content/FinalDS.xlsx'
df = None

# Check if file exists
if not os.path.exists(file_path):
    print(f"ERROR: File not found at {file_path}")
    print("Please upload your file to /content/ directory")
    print("\nTrying alternative paths...")

    # Try other common paths
    alternative_paths = [
        'FinalDS.xlsx',
        '/content/drive/MyDrive/FinalDS.xlsx',
        'FinalDS.csv'
    ]

    for alt_path in alternative_paths:
        if os.path.exists(alt_path):
            file_path = alt_path
            print(f"Found file at: {alt_path}")
            break
    else:
        raise FileNotFoundError("Dataset file not found. Please upload FinalDS.xlsx or FinalDS.csv")

# Try loading with different methods
print(f"\nAttempting to load: {file_path}")

try:
    # Method 1: Try as Excel with openpyxl
    print("Trying openpyxl engine...")
    df = pd.read_excel(file_path, engine='openpyxl')
    print("✓ Successfully loaded with openpyxl")
except Exception as e:
    print(f"✗ openpyxl failed: {str(e)[:100]}")

    try:
        # Method 2: Try as Excel with xlrd (for older .xls files)
        print("Trying xlrd engine...")
        df = pd.read_excel(file_path, engine='xlrd')
        print("✓ Successfully loaded with xlrd")
    except Exception as e:
        print(f"✗ xlrd failed: {str(e)[:100]}")

        try:
            # Method 3: Try as CSV
            print("Trying as CSV...")
            df = pd.read_csv(file_path)
            print("✓ Successfully loaded as CSV")
        except Exception as e:
            print(f"✗ CSV failed: {str(e)[:100]}")

            # Diagnose the file
            print("\n" + "="*80)
            print("FILE DIAGNOSIS")
            print("="*80)
            print(f"File size: {os.path.getsize(file_path)} bytes")

            with open(file_path, 'rb') as f:
                header = f.read(20)
                print(f"First 20 bytes (hex): {header.hex()}")
                print(f"First 20 bytes (repr): {repr(header)}")

                if header[:2] == b'PK':
                    print("\n→ File appears to be ZIP-based (xlsx format)")
                    print("  But pandas couldn't read it. File may be corrupted.")
                elif header[:2] == b'\xd0\xcf':
                    print("\n→ File appears to be older Excel format (.xls)")
                    print("  Try: pip install xlrd==1.2.0")
                else:
                    print("\n→ File format unknown or corrupted")

            print("\n" + "="*80)
            print("SOLUTION:")
            print("="*80)
            print("1. Re-upload your file")
            print("2. OR convert to CSV format:")
            print("   - Open in Excel/Google Sheets")
            print("   - File → Save As → CSV")
            print("   - Upload the CSV file")
            print("   - Change file_path to '/content/FinalDS.csv'")
            print("="*80)

            raise Exception("Could not load dataset. Please fix file format.")

# Verify dataset loaded successfully
if df is None:
    raise Exception("Failed to load dataset")

# Display initial info
print("\n" + "="*80)
print("DATASET INFORMATION")
print("="*80)
print(f"Dataset shape: {df.shape}")
print(f"Columns: {df.columns.tolist()}")
print(f"\nFirst few rows:")
print(df.head())

# Check for required columns
required_columns = ['TELUGU', 'ENGLISH']
missing_columns = [col for col in required_columns if col not in df.columns]

if missing_columns:
    print(f"\n⚠️  WARNING: Missing required columns: {missing_columns}")
    print(f"Available columns: {df.columns.tolist()}")
    print("\nPlease verify your dataset has 'TELUGU' and 'ENGLISH' columns")

    # Try to find similar columns
    print("\nSearching for similar column names...")
    for col in df.columns:
        if 'telugu' in col.lower() or 'tel' in col.lower():
            print(f"  Found potential Telugu column: '{col}'")
        if 'english' in col.lower() or 'eng' in col.lower():
            print(f"  Found potential English column: '{col}'")

    raise ValueError("Required columns not found in dataset")

# Select Telugu and English columns for translation
print("\n" + "="*80)
print("PREPARING DATA")
print("="*80)
df_filtered = df[['TELUGU', 'ENGLISH']].copy()
print(f"Before removing NaN: {df_filtered.shape}")

df_filtered = df_filtered.dropna().reset_index(drop=True)
print(f"After removing NaN: {df_filtered.shape}")

# Remove empty strings
df_filtered = df_filtered[(df_filtered['TELUGU'].str.strip() != '') &
                          (df_filtered['ENGLISH'].str.strip() != '')]
print(f"After removing empty strings: {df_filtered.shape}")

if len(df_filtered) == 0:
    raise ValueError("No valid data remaining after cleaning!")

# Prepare data for translation (Telugu -> English)
telugu_texts = df_filtered['TELUGU'].values
english_texts = df_filtered['ENGLISH'].values

print(f"\nNumber of translation pairs: {len(telugu_texts)}")
print(f"\nSample data:")
for i in range(min(3, len(telugu_texts))):
    print(f"\n{i+1}. Telugu: {telugu_texts[i][:100]}...")
    print(f"   English: {english_texts[i][:100]}...")

# Prepare target sequences with start and end tokens
english_texts_input = ['<start> ' + text for text in english_texts]
english_texts_output = [text + ' <end>' for text in english_texts]

# Tokenization
print("\n" + "="*80)
print("TOKENIZATION")
print("="*80)
max_vocab_size = 10000
max_len = 50

# Tokenizer for Telugu (source)
tokenizer_tel = Tokenizer(num_words=max_vocab_size, oov_token='<OOV>')
tokenizer_tel.fit_on_texts(telugu_texts)
tel_sequences = tokenizer_tel.texts_to_sequences(telugu_texts)
tel_padded = pad_sequences(tel_sequences, maxlen=max_len, padding='post')

# Tokenizer for English (target)
tokenizer_eng = Tokenizer(num_words=max_vocab_size, oov_token='<OOV>')
tokenizer_eng.fit_on_texts(english_texts_input + english_texts_output)
eng_input_sequences = tokenizer_eng.texts_to_sequences(english_texts_input)
eng_output_sequences = tokenizer_eng.texts_to_sequences(english_texts_output)

eng_input_padded = pad_sequences(eng_input_sequences, maxlen=max_len, padding='post')
eng_output_padded = pad_sequences(eng_output_sequences, maxlen=max_len, padding='post')

tel_vocab_size = min(len(tokenizer_tel.word_index) + 1, max_vocab_size)
eng_vocab_size = min(len(tokenizer_eng.word_index) + 1, max_vocab_size)

print(f"Telugu vocabulary size: {tel_vocab_size}")
print(f"English vocabulary size: {eng_vocab_size}")

# Split data
X_train_tel, X_test_tel, X_train_eng, X_test_eng, y_train, y_test = train_test_split(
    tel_padded, eng_input_padded, eng_output_padded, test_size=0.2, random_state=42
)

print(f"\nTraining samples: {len(X_train_tel)}")
print(f"Testing samples: {len(X_test_tel)}")

# Build LSTM Encoder-Decoder Model
print("\n" + "="*80)
print("BUILDING MODEL")
print("="*80)
embedding_dim = 256
lstm_units = 512

# Encoder
encoder_inputs = Input(shape=(max_len,), name='encoder_input')
encoder_embedding = Embedding(
    tel_vocab_size,
    embedding_dim,
    mask_zero=False,  # Changed to False to avoid masking issues
    name='encoder_embedding'
)(encoder_inputs)
encoder_lstm = LSTM(
    lstm_units,
    return_state=True,
    name='encoder_lstm'
)
encoder_outputs, state_h, state_c = encoder_lstm(encoder_embedding)
encoder_states = [state_h, state_c]

# Decoder
decoder_inputs = Input(shape=(max_len,), name='decoder_input')
decoder_embedding = Embedding(
    eng_vocab_size,
    embedding_dim,
    mask_zero=False,  # Changed to False to avoid masking issues
    name='decoder_embedding'
)(decoder_inputs)
decoder_lstm = LSTM(
    lstm_units,
    return_sequences=True,
    return_state=True,
    name='decoder_lstm'
)
decoder_outputs, _, _ = decoder_lstm(decoder_embedding, initial_state=encoder_states)
decoder_dense = Dense(eng_vocab_size, activation='softmax', name='decoder_dense')
decoder_outputs = decoder_dense(decoder_outputs)

# Model
model = Model([encoder_inputs, decoder_inputs], decoder_outputs)
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

print(model.summary())

# Prepare target data
y_train_expanded = np.expand_dims(y_train, -1)
y_test_expanded = np.expand_dims(y_test, -1)

# Train model
print("\n" + "="*80)
print("TRAINING MODEL")
print("="*80)
early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

history = model.fit(
    [X_train_tel, X_train_eng],
    y_train_expanded,
    validation_data=([X_test_tel, X_test_eng], y_test_expanded),
    epochs=50,
    batch_size=64,
    callbacks=[early_stop],
    verbose=1
)

# Build inference models
print("\n" + "="*80)
print("BUILDING INFERENCE MODELS")
print("="*80)
encoder_model = Model(encoder_inputs, encoder_states)

decoder_state_input_h = Input(shape=(lstm_units,), name='decoder_h_input')
decoder_state_input_c = Input(shape=(lstm_units,), name='decoder_c_input')
decoder_states_inputs = [decoder_state_input_h, decoder_state_input_c]

decoder_input_single = Input(shape=(1,), name='decoder_single_input')
decoder_embedding_inf = Embedding(
    eng_vocab_size,
    embedding_dim,
    mask_zero=False,
    name='decoder_embedding_inf'
)(decoder_input_single)
decoder_outputs_inf, state_h_inf, state_c_inf = decoder_lstm(
    decoder_embedding_inf, initial_state=decoder_states_inputs
)
decoder_states_inf = [state_h_inf, state_c_inf]
decoder_outputs_inf = decoder_dense(decoder_outputs_inf)

decoder_model = Model(
    [decoder_input_single] + decoder_states_inputs,
    [decoder_outputs_inf] + decoder_states_inf
)

# Decode sequence function
def greedy_decode(input_seq, max_len_decode=max_len):
    states_value = encoder_model.predict(input_seq, verbose=0)

    target_seq = np.zeros((1, 1))
    target_seq[0, 0] = tokenizer_eng.word_index.get('<start>', 1)

    stop_condition = False
    decoded_sentence = []

    while not stop_condition:
        output_tokens, h, c = decoder_model.predict([target_seq] + states_value, verbose=0)

        sampled_token_index = np.argmax(output_tokens[0, -1, :])
        sampled_word = tokenizer_eng.index_word.get(sampled_token_index)

        if sampled_word == '<end>' or len(decoded_sentence) > max_len_decode or sampled_word is None:
            stop_condition = True
        elif sampled_word != '<start>':
            decoded_sentence.append(sampled_word)

        target_seq = np.zeros((1, 1))
        target_seq[0, 0] = sampled_token_index

        states_value = [h, c]

    return ' '.join(decoded_sentence)

# Generate predictions on test set
print("\n" + "="*80)
print("GENERATING PREDICTIONS")
print("="*80)
predictions = []
references = []

num_samples = min(2000, len(X_test_tel))
print(f"Processing {num_samples} samples for evaluation...")

for i in range(num_samples):
    input_seq = X_test_tel[i:i+1]
    decoded = greedy_decode(input_seq)
    predictions.append(decoded)

    # Get reference
    ref_indices = y_test[i]
    ref_words = []
    for idx in ref_indices:
        if idx != 0:
            word = tokenizer_eng.index_word.get(idx)
            if word and word not in ['<start>', '<end>', '<OOV>']:
                ref_words.append(word)
    references.append(' '.join(ref_words))

    if (i + 1) % 100 == 0:
        print(f"Processed {i+1}/{num_samples} samples")

print(f"\nTotal predictions: {len(predictions)}")

# Calculate BLEU scores
print("\n" + "="*80)
print("CALCULATING BLEU SCORES")
print("="*80)
bleu_scores = []
smoothing = SmoothingFunction().method1

for pred, ref in zip(predictions, references):
    pred_tokens = pred.split()
    ref_tokens = [ref.split()]

    if len(pred_tokens) > 0 and len(ref_tokens[0]) > 0:
        score = sentence_bleu(ref_tokens, pred_tokens, smoothing_function=smoothing)
        bleu_scores.append(score * 100)
    else:
        bleu_scores.append(0)

avg_bleu = np.mean(bleu_scores)
print(f"Average BLEU Score: {avg_bleu:.2f}")
print(f"Min BLEU Score: {np.min(bleu_scores):.2f}")
print(f"Max BLEU Score: {np.max(bleu_scores):.2f}")
print(f"Std Dev BLEU Score: {np.std(bleu_scores):.2f}")

# Calculate ROUGE scores
print("\n" + "="*80)
print("CALCULATING ROUGE SCORES")
print("="*80)
rouge1_scores = []
rouge2_scores = []
rougel_scores = []

for pred, ref in zip(predictions, references):
    if len(pred.strip()) > 0 and len(ref.strip()) > 0:
        r1 = calculate_rouge_n(ref, pred, n=1) * 100
        r2 = calculate_rouge_n(ref, pred, n=2) * 100
        rl = calculate_rouge_l(ref, pred) * 100

        rouge1_scores.append(r1)
        rouge2_scores.append(r2)
        rougel_scores.append(rl)
    else:
        rouge1_scores.append(0)
        rouge2_scores.append(0)
        rougel_scores.append(0)

avg_rouge1 = np.mean(rouge1_scores)
avg_rouge2 = np.mean(rouge2_scores)
avg_rougel = np.mean(rougel_scores)

print(f"Average ROUGE-1 F1 Score: {avg_rouge1:.2f}")
print(f"Average ROUGE-2 F1 Score: {avg_rouge2:.2f}")
print(f"Average ROUGE-L F1 Score: {avg_rougel:.2f}")

# Print sample translations
print("\n" + "="*80)
print("SAMPLE TRANSLATIONS")
print("="*80)
for i in range(min(10, len(predictions))):
    print(f"\nSample {i+1}:")
    print(f"Reference: {references[i]}")
    print(f"Predicted: {predictions[i]}")
    print(f"BLEU: {bleu_scores[i]:.2f}, ROUGE-1: {rouge1_scores[i]:.2f}, ROUGE-2: {rouge2_scores[i]:.2f}, ROUGE-L: {rougel_scores[i]:.2f}")
    print("-" * 80)

# Plot BLEU scores
print("\nGenerating BLEU score visualizations...")
plt.figure(figsize=(16, 6))

plt.subplot(1, 3, 1)
plt.hist(bleu_scores, bins=50, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(avg_bleu, color='red', linestyle='--', linewidth=2, label=f'Mean: {avg_bleu:.2f}')
plt.xlabel('BLEU Score', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title(f'Distribution of BLEU Scores (n={num_samples})', fontsize=14, fontweight='bold')
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(1, 3, 2)
sample_indices = list(range(min(50, len(bleu_scores))))
sample_bleu = [bleu_scores[i] for i in sample_indices]
plt.bar(sample_indices, sample_bleu, color='steelblue', alpha=0.7)
plt.axhline(avg_bleu, color='red', linestyle='--', linewidth=2, label=f'Average: {avg_bleu:.2f}')
plt.xlabel('Sample Index', fontsize=12)
plt.ylabel('BLEU Score', fontsize=12)
plt.title('BLEU Scores for Sample Translations', fontsize=14, fontweight='bold')
plt.legend()
plt.grid(True, alpha=0.3, axis='y')

plt.subplot(1, 3, 3)
plt.boxplot(bleu_scores, vert=True)
plt.ylabel('BLEU Score', fontsize=12)
plt.title('BLEU Score Box Plot', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.savefig('bleu_scores_2000.png', dpi=300, bbox_inches='tight')
plt.show()
print("BLEU scores plot saved as 'bleu_scores_2000.png'")

# Plot ROUGE scores
print("\nGenerating ROUGE score visualizations...")
plt.figure(figsize=(16, 12))

plt.subplot(2, 3, 1)
scores_df = pd.DataFrame({
    'Metric': ['ROUGE-1', 'ROUGE-2', 'ROUGE-L'],
    'Score': [avg_rouge1, avg_rouge2, avg_rougel]
})
bars = plt.bar(scores_df['Metric'], scores_df['Score'], color=['#FF6B6B', '#4ECDC4', '#45B7D1'], alpha=0.8)
plt.ylabel('Score', fontsize=12)
plt.title(f'Average ROUGE Scores (n={num_samples})', fontsize=14, fontweight='bold')
plt.ylim(0, 100)
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{height:.2f}', ha='center', va='bottom', fontsize=10, fontweight='bold')
plt.grid(True, alpha=0.3, axis='y')

plt.subplot(2, 3, 2)
sample_indices = list(range(min(50, len(rouge1_scores))))
plt.plot(sample_indices, [rouge1_scores[i] for i in sample_indices], marker='o', label='ROUGE-1', linewidth=2, markersize=4)
plt.plot(sample_indices, [rouge2_scores[i] for i in sample_indices], marker='s', label='ROUGE-2', linewidth=2, markersize=4)
plt.plot(sample_indices, [rougel_scores[i] for i in sample_indices], marker='^', label='ROUGE-L', linewidth=2, markersize=4)
plt.xlabel('Sample Index', fontsize=12)
plt.ylabel('Score', fontsize=12)
plt.title('ROUGE Scores per Sample', fontsize=14, fontweight='bold')
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(2, 3, 3)
plt.hist(rouge1_scores, bins=50, color='#FF6B6B', alpha=0.6, edgecolor='black')
plt.axvline(avg_rouge1, color='darkred', linestyle='--', linewidth=2, label=f'Mean: {avg_rouge1:.2f}')
plt.xlabel('Score', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title('ROUGE-1 Distribution', fontsize=14, fontweight='bold')
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(2, 3, 4)
plt.hist(rouge2_scores, bins=50, color='#4ECDC4', alpha=0.6, edgecolor='black')
plt.axvline(avg_rouge2, color='darkblue', linestyle='--', linewidth=2, label=f'Mean: {avg_rouge2:.2f}')
plt.xlabel('Score', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title('ROUGE-2 Distribution', fontsize=14, fontweight='bold')
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(2, 3, 5)
plt.hist(rougel_scores, bins=50, color='#45B7D1', alpha=0.6, edgecolor='black')
plt.axvline(avg_rougel, color='darkgreen', linestyle='--', linewidth=2, label=f'Mean: {avg_rougel:.2f}')
plt.xlabel('Score', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title('ROUGE-L Distribution', fontsize=14, fontweight='bold')
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(2, 3, 6)
comparison_data = pd.DataFrame({
    'Sample': sample_indices[:20],
    'ROUGE-1': [rouge1_scores[i] for i in sample_indices[:20]],
    'ROUGE-2': [rouge2_scores[i] for i in sample_indices[:20]],
    'ROUGE-L': [rougel_scores[i] for i in sample_indices[:20]]
})
x = np.arange(len(comparison_data))
width = 0.25
plt.bar(x - width, comparison_data['ROUGE-1'], width, label='ROUGE-1', color='#FF6B6B', alpha=0.8)
plt.bar(x, comparison_data['ROUGE-2'], width, label='ROUGE-2', color='#4ECDC4', alpha=0.8)
plt.bar(x + width, comparison_data['ROUGE-L'], width, label='ROUGE-L', color='#45B7D1', alpha=0.8)
plt.xlabel('Sample Index', fontsize=12)
plt.ylabel('Score', fontsize=12)
plt.title('ROUGE Scores Comparison', fontsize=14, fontweight='bold')
plt.xticks(x, comparison_data['Sample'], rotation=45)
plt.legend()
plt.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.savefig('rouge_scores_2000.png', dpi=300, bbox_inches='tight')
plt.show()
print("ROUGE scores plot saved as 'rouge_scores_2000.png'")

# Combined metrics summary plot
print("\nGenerating combined metrics visualization...")
plt.figure(figsize=(10, 6))
all_metrics = {
    'BLEU': avg_bleu,
    'ROUGE-1': avg_rouge1,
    'ROUGE-2': avg_rouge2,
    'ROUGE-L': avg_rougel
}
colors = ['#3498db', '#e74c3c', '#2ecc71', '#f39c12']
bars = plt.bar(all_metrics.keys(), all_metrics.values(), color=colors, alpha=0.8, edgecolor='black', linewidth=2)
plt.ylabel('Score', fontsize=12)
plt.title(f'All Evaluation Metrics Summary (n={num_samples})', fontsize=14, fontweight='bold')
plt.ylim(0, 100)
for bar, (metric, value) in zip(bars, all_metrics.items()):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{value:.2f}', ha='center', va='bottom', fontsize=11, fontweight='bold')
plt.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.savefig('all_metrics_summary_2000.png', dpi=300, bbox_inches='tight')
plt.show()
print("Combined metrics plot saved as 'all_metrics_summary_2000.png'")

# Final summary
print("\n" + "="*80)
print("FINAL RESULTS SUMMARY")
print("="*80)
print(f"\nEvaluation samples: {num_samples}")
print(f"\nBLEU Score Statistics:")
print(f"  Average: {avg_bleu:.2f}")
print(f"  Min: {np.min(bleu_scores):.2f}")
print(f"  Max: {np.max(bleu_scores):.2f}")
print(f"  Std Dev: {np.std(bleu_scores):.2f}")

print(f"\nROUGE Score Statistics:")
print(f"  ROUGE-1 Average: {avg_rouge1:.2f}")
print(f"  ROUGE-2 Average: {avg_rouge2:.2f}")
print(f"  ROUGE-L Average: {avg_rougel:.2f}")
print("="*80)

# Save model
print("\nSaving model...")
try:
    model.save('lstm_translation_model.h5')
    print("✓ Model saved as 'lstm_translation_model.h5'")
except Exception as e:
    print(f"⚠️  Warning: Could not save model: {e}")

print("\n" + "="*80)
print("TRAINING COMPLETE!")
print("="*80)
```
